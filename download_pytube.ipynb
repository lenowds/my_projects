````python
# Vamos instânciar o pacote Youtube da biblioteca pytube, e a biblioteca IPython
# Let's go instance the package Youtube from pytube library and IPython library
from pytube import YouTube
import IPython
````

````python
"""
O pacote Youtube permite registrar funções de retorno durante o progresso e após o download concluído, para mostrar um gif durante o download por exemplo.

Estou usando apenas a função de retorno após concluído o download

The Youtube package allow using on progress callback function and complete callback funcation. This is allow displaying gif image for example.

I'am using only complete callback function
"""

def download_concluido(stream, file_path):
    print(f"Download {stream.mime_type} guardado em: {file_path}")
        
````

````python


video = YouTube('https://www.youtube.com/watch?v=mFsHA1WDm0g',
                on_complete_callback=download_concluido)
````

````python
"""
Vou carregar a imagem de capa do vídeo que optei usar neste projeto

The image thumbnail from video
"""
IPython.display.Image(video.thumbnail_url)
````

````python
# Vou baixar agora o vídeo diretamente do youtube
# Now download the video file from youtube 
print(f"Arquivo de vídeo [{video.title}] será baixado do YouTube.")
video.streams.first().download();
````

Até agora estávamos baixando o vídeo completo do youtube, e que tal ao invés de baixar o vídeo completo baixarmos apenas a faixa de áudio?

Da biblioteca **pytube** vamos importar apenas o pacote *Playlist*. Este pacote permite trabalhar e manipular playlist no Youtube.





We up to now download the full video from youtube but now let’s go download only the track audio

We import from **pytube** a *Playlist* library. It's package allow working with Playlist on youtube.

````python
# import pytube library
from pytube import Playlist
````

Uma **Playlist** pode ser criada no **Youtube** por qualquer usuário que tenha uma conta no [google](https://www.youtube.com/). Vídeos são adicionados na biblioteca permitindo o usuário personalizar com suas coletâneas favoritas e tocar sua sequência de vídeos nos melhores momentos.

A **Playlist** can create on **Youtube** by any users with an account on [Google](https://www.youtube.com/). The Videos are add library allowing the user personalize and play your videos on best moments.


````python
# Vamos instânciar o objeto Playlist
# Let's go instance the Playlist object
library = Playlist('https://www.youtube.com/playlist?list=OLAK5uy_lbQx-BZ7-nxpmQkH7tF82df4a_rTvfiYw')
````
